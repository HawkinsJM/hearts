```python
#this function takes a single message entry from the chat and separates out the time, user, text and emoji into a dict
def parse_message(message):
    
    #The first p is always date/time, the second one contains the username/message
    paras = message.find_all("p")
    
    #This pulls the date as a string
    date = paras[0].text
    
    #this checks to see if there is a tag for the span containing the username, then grabs it apropriately, and removes the span element containgin the user name
    name_tag = message("span", "c_username")  
    if len(name_tag) > 0:
        user = name_tag[0].text[:-2]
        name_tag[0].extract()
    else:
        user = paras[1].span.text[:-2]
        paras[1].span.extract()
        
    #pull emoji
    emojis = paras[1].find_all("div", "smiley")
    emoji_list = []
    for e in emojis:
        paths = [path for path in emoji_dict.keys() if path in str(e)]
        if len(paths) > 0:
            for path in paths:
                emoji_list.append(emoji_dict[path])
        
        else:
            print "emoji not defined"
            print e        
        
    text = paras[1].text
    
    mdict = {"user":user, "date": date, "text": text, "emojis": emoji_list}
    return mdict
```

```python
#This function counts the heart emoji's present in a comment. 
#Scans the text for emojis and also looks in the emoji colum for heart emoji.
def count_hearts(row):
    emojis = row.emojis
    text = row.text

    hearts = 0
    
    #count the emoji hearts
    heart_emojis = ["heart", "broken_heart"]
    if type(emojis) == list:
        for e in emojis:
            if e in heart_emojis:
                hearts += 1
            
    #count the text hearts
    hs = ["<","", "\*h","\*H", "", "","","HEART", ""]
    if isinstance(text, unicode):
        text = text.encode('utf8','ignore')
        for h in hs:
            found  = len(re.findall(h,text))
            hearts += found
    return hearts

# #these test a specified entry.
# row = mdf.iloc[2587]
# print count_hearts(row)
```

```python
#Runs through all the rows parsing them.
#Returns a dataframe
messages_dicts = []
for message in messages_html:
    mdict = parse_message(message)
    messages_dicts.append(mdict)

mdf = pd.DataFrame(messages_dicts)
```

```python
#replace empty lists with nan
mdf.emojis = mdf.emojis.apply(lambda y: np.nan if len(y)==0 else y)
#replace blank with nan
mdf.replace(r'^$', np.nan, regex=True, inplace = True)
mdf.replace(r'^\s+$', np.nan, regex=True, inplace = True)
#convert date to date time and make it the index
mdf['date'] =  pd.to_datetime(mdf['date'], format='%a %b %d, %H:%M:%S%p')
mdf['date'] = mdf["date"].apply(lambda dt: dt.replace(year=2017))
#make hearts column
mdf['hearts'] = mdf.apply(lambda row: int(count_hearts(row)), axis =1)
```
